```python
import time
import matplotlib.pyplot as plt
from math import isclose

def compare(problems, function_a, function_b, title, x):
    # Listas para almacenar tiempos y resultados
    times_a = []
    times_b = []
    results_a = []
    results_b = []

    for problem in problems:
        start_time = time.time()
        result_a = function_a(problem)
        elapsed_time_a = time.time() - start_time
        start_time = time.time()
        result_b = function_b(problem)
        elapsed_time_b = time.time() - start_time
        
        # Almacenar resultados y tiempos
        results_a.append(result_a)
        results_b.append(result_b)
        times_a.append(elapsed_time_a)
        times_b.append(elapsed_time_b)

        if isinstance(result_a['result'],float) or isinstance(result_b['result'], float):
            # Usa una tolerancia para la comparación de floats
            if not isclose(result_a['result'], result_b['result'], rel_tol=1e-1, abs_tol=1e-1):
                print(f'Error:')
                print(problem)
                print("Solución CSP:")
                print(result_a)
                print("Solución IP:")
                print(result_b)

        elif result_a['result'] != result_b['result']:
            print(f'Error:')
            print(problem)
            print("Solución CSP:")
            print(result_a)
            print("Solución IP:")
            print(result_b)

    # Graficar los tiempos
    plt.plot(range(len(problems)), times_a, label='CSP')
    plt.plot(range(len(problems)), times_b, label='IP')

    plt.title(title)
    plt.xlabel(x)
    plt.ylabel('Tiempo (segundos)')
    plt.legend()
    plt.grid()
    plt.show()
```

## Knapsack problem

```python
from Knapsack.Knapsack import problem_builder
from Knapsack.CSP import solve as CSP_solve
from Knapsack.IP import solve as IP_solve

# Generar una lista de problemas
problems = [problem_builder(i+1,100) for i in range(100)]
compare(problems, CSP_solve, IP_solve, "Problema de la mochila", "Cantidad de objetos")
```
